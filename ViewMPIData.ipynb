# View MPI Data

View MPI data in all storage locations

- NoSQL DB

- SQL DB

```python
from db import get_session, get_mongo
import json

def get_mongo_record(mpi: str) -> tuple:
    db = get_mongo()
    x = db.raw.find_one({'mpi': mpi})
    x['_id'] = str(x['_id'])
    return json.dumps(x, indent=2), x

def get_sql_records(mpi: str) -> tuple:
    x = {'mpi': mpi, 'vectors':[]}
    with get_session() as session:
        res = session.execute(f"SELECT * FROM mpi_vectors WHERE mpi='{mpi}'").fetchall()
        for r in res:
            x['vectors'].append(list(r))
    return json.dumps(x, indent=2), x
            
```

```python
print(get_mongo_record('15918315-13223288-3889668-8685359')[0])
```

```python
print(get_sql_records('15918315-13223288-3889668-8685359')[0])
```
